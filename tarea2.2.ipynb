### Función esfera (unimodal)

Ecuación:

$f(x) = \sum_{i=1}^{n} {x_i^2}$

Forma desarollada para 3 dimensiones:
  
  $z = x^2 + y^2$

```python
# Esfera (unimodal)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import *

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.set_title("Sphere")

X = linspace(-10, 10, 50)
Y = linspace(-10, 10, 50)

X, Y = meshgrid(X, Y)

def f(x, y):
  return x**2 + y**2

ax.plot_surface(X, Y, f(X, Y), cmap = 'viridis', edgecolor = 'none')

plt.show()
```

### Step function

Ecuación: 

$f(x) = \sum_{i=1}^{n} {\lfloor|x_i|\rfloor}$

```python
# Step function
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import *

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.set_title("Step function")

X = linspace(-100, 100, 50)
Y = linspace(-100, 100, 50)

X, Y = meshgrid(X, Y)

def f(x, y):
  return floor(abs(x)) 

ax.plot_surface(X, Y, f(X, Y), cmap = 'viridis', edgecolor = 'none')

plt.show()
```

### Michalewicz function

Ecuación:

$f(x) = -\sum_{i=1}^{n} {sin(x_i) * sin^{2m}{(\frac{ix_i^2}{\pi})}}$

```python
# Michalewicz function
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import *

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.set_title("Michalewicz function")

d = 50
m = 10
X = linspace(0, pi, d)
Y = linspace(0, pi, d)

X, Y = meshgrid(X, Y)

def f(x, i):
  return -sin(x) * sin(i * x**2 / pi) ** (2 * m)

ax.plot_surface(X, Y, f(X, Y), cmap = 'viridis', edgecolor = 'none')

plt.show()
```

### Eggholder function

Ecuación:

$f(x, y) = -(y + 47)sin(\sqrt{|\frac{x}{2} + y + 47|}) - x sin(\sqrt{|x - y - 47|})$

```python
# Eggholder function
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import *

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.set_title("Eggholder function")

X = linspace(-512, 512, 50)
Y = linspace(-512, 512, 50)
X, Y = meshgrid(X, Y)

def f(x, y): 
  yPlus47 = y + 47
  return -yPlus47 * sin(sqrt(abs(x / 2 + yPlus47))) - x * sin(sqrt(abs(x - yPlus47)))

# print(f(512, 404.2319)) # -956.6407

ax.plot_surface(X, Y, f(X, Y), cmap = 'viridis', edgecolor = 'none')

plt.show()
```

### Weierstrass function

Ecuación:

$f(x) = \sum_{i=1}^{n} {[ \sum_{k=0}^{kmax}{a^k cos(2 \pi b^k (x_i + 0.5))} − n \sum_{k=0}^{kmax} {a^k cos(\pi b^k)}] } $

Donde $0<a<1$, $b$ es un entero positivo impar, y $ab > 1 + \frac{3}{2}\pi$



```python
# Weierstrass function
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from numpy import *
import random

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.set_title("Weierstrass function")

a = random.uniform(0, 1)
b = random.randint(1, 100)
n = 50
kmax = 20

X = linspace(-10, 10, N)
Y = linspace(-10, 10, N)

X, Y = meshgrid(X, Y)

def f(x, y):
  add = 0
  rem = 0
  for k in range(0, kmax + 1):
    add += a**k * cos(2 * pi * b**k * (x + 0.5))
    rem += a**k * cos(pi * b**k)
  return add - n * rem

ax.plot_surface(X, Y, f(X, Y), cmap = 'viridis', edgecolor = 'none')

plt.show()
```
